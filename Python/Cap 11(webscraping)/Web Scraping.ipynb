### Salvando arquivos baixaxdos no disco rígido 

```python
import requests

res = requests.get('https://automatetheboringstuff.com/files/rj.txt')
#Faz download da pagina web

res.raise_for_status()#Verifica se a pagina web tem alguum problema
#Caso tenha, retorna um erro, caso contrário nada

playFile = open('RomeoAndJuliet.txt', 'wb')
#Cria um arquivo txt em modo wb(retornará um binário)
#Caso contrário, poderá vir com erros

for chunk in res.iter_content(100000):#res.iter_content
#Retorna uma quantidade de bytes, dividido em 'porçoes'
#Passar 100000, significa que ele passará 100 mil bytes por vez
    playFile.write(chunk)#Escreverá todas palavras da pagina no arquivo
    

playFile.close()

''' 
            Método para fazer download de páginas e gravar no disco
1. Chame requests.get() para fazer download do arquivo.
2. Chame open() com 'wb' para criar um novo arquivo em modo de escrita
binária.
3. Crie um loop com o método iter_content() do objeto Response.
4. Chame write() a cada iteração para gravar o conteúdo no arquivo.
5. Chame close() para fechar o arquivo.

'''





```

## Projeto: Downloading de todas as tirinhas XKCD


```python
import os

url = 'https://xkcd.com' #url inicial
os.makedirs('Tirinhas', exist_ok=True) #Cria uma pasta xkcd e amarzena lá
while not url.endswith("#"):
    print("Downloading page %s..." % url)
    res = requests.get(url)#Faz o request para fazer download da pagina
    res.raise_for_status()#Verifica se tem erro
    soup = bs4.BeautifulSoup(res.text)#Passa o texto do download da pagina(o html) para a váriavel
    comicElem = soup.select("#comic img") #No elemento de id = comic, pegar a img
    if comicElem == []:
        print("Could not find comic image") #Se não houver imagem, printará isso
        prevLink = soup.select('a[rel="prev"]')[0] #Vai selecionar no elemento a, elemento 'rel' igual a classe'prev'
        url = 'http://xkcd.com' + prevLink.get('href')#Pegando a referencia do botão prev, fará um parse com a url
    else:
        comicUrl = 'http:' + comicElem[0].get('src')#Pega o url do elemento img
        
        print('Downloading image %s' %(comicUrl))#Printará download da imagem
        res = requests.get(comicUrl)#Vai fazer um get para pegar a página que tem a imagem
        res.raise_for_status()#Verifica se tem erro
        imageFile = open(os.path.join('Tirinhas', os.path.basename(comicUrl)), 'wb')#Vai abrir a pasta tirinhas, o arquivo
        #Com o base nome do path comic url(Pegará só o nome da tirinha)
        for chunk in res.iter_content(100000):#Escreverá a imagem em tamanho de bits nesse caso 100000 por vez
            imageFile.write(chunk)#Baixará a imagem 
        imageFile.close()#Fecha a imagem
        
        prevLink = soup.select('a[rel="prev"]')[0]#Vai selecionar no elemento a, elemento 'rel' igual a classe'prev'
        url = 'http://xkcd.com' + prevLink.get('href'))#Pegando a referencia do botão prev, fará um parse com a url
        #Para sempre ficar voltando a pag
        

print("Done")
```
